```python

#importing libraries

from nptdms import TdmsFile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
from scipy import signal
from scipy import io
from scipy import sparse
from scipy.sparse.linalg import spsolve
from scipy.linalg import solveh_banded
from numpy import array, savetxt
%matplotlib inline

#Loading mat file of AsLS signals (BL and TN)

num_to_plot = 48

for i in range(1,num_to_plot):
    
    curr_val_dict = io.loadmat("AsLS_treated_signal_C10_dNTP (%s).mat"%i)

    time_curr_val = curr_val_dict['array']

    time_val = time_curr_val[:,0]
    curr_val = time_curr_val[:,1]

    #Plot the actual data point

    plt.plot(time_val,curr_val)
    plt.show()
    

    #Inverting current arrays

    inv_curr_val = curr_val * -1.0

    #plot the inverted data points

    plt.plot(time_val,inv_curr_val)
    plt.show() 
    

    #Empty list to store time and curr values > 10pA 

    updated_time_val = []
    updated_curr_val = []

    for k in range(inv_curr_val.shape[0]):
        if inv_curr_val[k] > 0.01:
            updated_curr_val.append(inv_curr_val[k])
            updated_time_val.append(time_val[k])



    print("length of updated time list", len(updated_time_val))
    print("length of updated curr list", len(updated_curr_val))


    updated_time_val = np.asarray(updated_time_val)
    updated_curr_val = np.asarray(updated_curr_val)


    plt.plot(updated_time_val,updated_curr_val)
    plt.savefig("BL_C10_dNTP (%s).png"%i, bbox_inches="tight", dpi=600)
    plt.show()
    
    


    #histogram

    plt.hist(updated_curr_val, bins='auto',edgecolor="k")
    plt.savefig("BL_histogram_C10_dNTP (%s).png"%i, bbox_inches="tight", dpi=600) 
    plt.show()
    
    

    #Numpy histogram to get bin centres and counts and saving it

    hist, bin_edges = np.histogram(updated_curr_val, bins='auto')

    print("hist", hist)
    print("bin_edges",bin_edges)

    avg_val=[]
    for j in range(bin_edges.shape[0]-1):
        tmp= (bin_edges[j+1]+bin_edges[j])/2.0
        avg_val.append(tmp)

    print(len(avg_val))
    avg_val=np.asarray(avg_val)
    print(avg_val.shape)
    print(avg_val)

    data = np.column_stack([avg_val, hist])
    print(data.shape)

    np.savetxt("BL_histogram_C10_dNTP (%s).csv"%i, data , delimiter=",",  header= " current, count")


```
